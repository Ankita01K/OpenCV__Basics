Blurring

```python
img2=cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)     #This line of code converts the input image img1 from the BGR color space to the HSV (Hue, Saturation, Value)
                                              # color space using the cv2.cvtColor function.
figure_size=9
result=cv2.blur(img2,(figure_size,figure_size))  #The code applies a blurring operation to the HSV image (img2) using the cv2.blur function.
                                                 #The blurring is achieved by using a kernel of size (figure_size, figure_size).
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(cv2.cvtColor(img1,cv2.COLOR_HSV2RGB)),plt.title("original")#This line sets up the first subplot (1 row, 2 columns), where it displays the original image (img1)
                                                                                      #converted from HSV to RGB using cv2.cvtColor and sets the title as "original
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(cv2.cvtColor(result,cv2.COLOR_HSV2RGB)),plt.title("Blurred")
plt.show
```

the code converts an image from the BGR color space to HSV, applies blurring, and then displays the original and blurred images side by side using Matplotlib subplots.

Zoom The Image

```python
(height,width)=img1.shape[0:2]
zoom=cv2.resize(img,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom=cv2.resize(img,(int(width*2),int(height*2)),interpolation=cv2.INTER_CUBIC)#The interpolation method used during resizing. Here,
                                                                               #cubic interpolation is employed for smoother results.
plt.imshow(zoom)
```

the code takes the input image, resizes it to double its width and height using cubic interpolation, and then displays the zoomed image using Matplotlib.







Rotate the Image

```python
m=cv2.getRotationMatrix2D((width/2,height/2),30 ,1)#30: The rotation angle in degrees. The image will be rotated by 30 degrees.
                                                   #1: The scaling factor. Here, it is set to 1, indicating no scaling.
rotate=cv2.warpAffine(img1,m,(height,width))
plt.imshow(rotate)
```

the code rotates the input image (img1) by 30 degrees around its center using a 2D rotation matrix and then displays the rotated image using Matplotlib.

Translate The Image

```python
m1=np.float32([[1,0,50],[0,1,100]])  # The first row [1, 0, 50] represents the transformation in the x-direction. Here, it indicates a translation of 50 pixels to the right.
                                      # The second row [0, 1, 100] represents the transformation in the y-direction. It indicates a translation of 100 pixels downwards.
dst=cv2.warpAffine(img,m1,(height,width))
plt.imshow(dst)
```

code applies an affine transformation to the input image (img) by translating it 50 pixels to the right and 100 pixels downwards. The transformed image is then displayed using Matplotlib.
